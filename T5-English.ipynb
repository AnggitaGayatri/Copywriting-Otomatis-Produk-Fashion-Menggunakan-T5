# KEYWORDS EXTRACTION

## TF-IDF

```python
import re, os, string
import pandas as pd

# Scikit-learn importings
from sklearn.feature_extraction.text import TfidfVectorizer

import nltk
nltk.download('punkt')
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('indonesian')
```

```python
# Fungsi untuk menghapus tanda baca
def clean_text(text):
    return ''.join([char for char in text if char not in string.punctuation])

# Membersihkan teks dalam kolom 'translated_description'
df['Cleaned_Description'] = df['Translated_Deskripsi'].apply(lambda x: clean_text(x.lower()))

# Fungsi untuk mengekstrak kata kunci dari teks menggunakan TF-IDF
def sort_coo(coo_matrix):
    """Sort a dict with highest score"""
    tuples = zip(coo_matrix.col, coo_matrix.data)
    return sorted(tuples, key=lambda x: (x[1], x[0]), reverse=True)

def extract_topn_from_vector(feature_names, sorted_items, topn=10):
    """get the feature names and tf-idf score of top n items"""
    sorted_items = sorted_items[:topn]
    score_vals = []
    feature_vals = []
    for idx, score in sorted_items:
        score_vals.append(round(score, 3))
        feature_vals.append(feature_names[idx])

    results = {}
    for idx in range(len(feature_vals)):
        results[feature_vals[idx]] = score_vals[idx]
    return results

def get_keywords(vectorizer, feature_names, doc, top_k=5):
    """Return top k keywords from a doc using TF-IDF method"""
    tf_idf_vector = vectorizer.transform([doc])
    sorted_items = sort_coo(tf_idf_vector.tocoo())
    keywords = extract_topn_from_vector(feature_names, sorted_items, top_k)
    return list(keywords.keys())

# Membuat vektorizer TF-IDF
vectorizer = TfidfVectorizer(stop_words=stop_words, smooth_idf=True, use_idf=True)

# Menyusun daftar deskripsi untuk diproses
corpora = df['Cleaned_Description'].tolist()

# Fit TF-IDF vectorizer
vectorizer.fit_transform(corpora)

# Menyimpan nama fitur dari vektorizer
feature_names = vectorizer.get_feature_names_out()

# Ekstraksi kata kunci
result = []
for doc in corpora:
    keywords = get_keywords(vectorizer, feature_names, doc, top_k=5)
    result.append({
        'description': doc,
        'keywords': ", ".join(keywords)
    })

# Membuat DataFrame dari hasil
final = pd.DataFrame(result)

# Menyimpan hasil ke file CSV
final.to_csv('translated_with_keywords.csv', index=False)

# Menampilkan hasil
print(final.head())
```

## Dataset Preparation

```python
class NewsSummaryDataset(Dataset):
    def __init__(self, data, tokenizer, text_max_token_len=128, summary_max_token_len=512):
        self.data = data
        self.tokenizer = tokenizer
        self.text_max_token_len = text_max_token_len
        self.summary_max_token_len = summary_max_token_len

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        data_row = self.data.iloc[index]
        text = data_row["keywords"]
        summary = data_row["description"]

        text_encoding = self.tokenizer(
            text, max_length=self.text_max_token_len, padding="max_length",
            truncation=True, return_attention_mask=True, return_tensors="pt"
        )

        summary_encoding = self.tokenizer(
            summary, max_length=self.summary_max_token_len, padding="max_length",
            truncation=True, return_attention_mask=True, return_tensors="pt"
        )

        labels = summary_encoding["input_ids"]
        labels[labels == 0] = -100

        return dict(
            text=text,
            summary=summary,
            text_input_ids=text_encoding["input_ids"].flatten(),
            text_attention_mask=text_encoding["attention_mask"].flatten(),
            labels=labels.flatten(),
            labels_attention_mask=summary_encoding["attention_mask"].flatten()
        )
```

## Data Module

```python
class NewsSummaryDataModule(pl.LightningDataModule):
    def __init__(self, train_df, test_df, tokenizer, batch_size=4, text_max_token_len=128, summary_max_token_len=512):
        super().__init__()
        self.train_df = train_df
        self.test_df = test_df
        self.tokenizer = tokenizer
        self.batch_size = batch_size
        self.text_max_token_len = text_max_token_len
        self.summary_max_token_len = summary_max_token_len

    def setup(self, stage=None):
        self.train_dataset = NewsSummaryDataset(self.train_df, self.tokenizer, self.text_max_token_len, self.summary_max_token_len)
        self.test_dataset = NewsSummaryDataset(self.test_df, self.tokenizer, self.text_max_token_len, self.summary_max_token_len)

    def train_dataloader(self):
        return DataLoader(self.train_dataset, batch_size=self.batch_size, shuffle=True, num_workers=2)

    def val_dataloader(self):
        return DataLoader(self.test_dataset, batch_size=self.batch_size, shuffle=False, num_workers=2)

    def test_dataloader(self):
        return DataLoader(self.test_dataset, batch_size=self.batch_size, shuffle=False, num_workers=2)
```

## Tokenizer

```python
MODEL_NAME = "t5-base"
tokenizer = T5Tokenizer.from_pretrained(MODEL_NAME)
```

# Model

```python
# Model
class NewsSummaryModel(pl.LightningModule):
    def __init__(self):
        super().__init__()
        self.model = T5ForConditionalGeneration.from_pretrained(MODEL_NAME, return_dict=True)

    def forward(self, input_ids, attention_mask, decoder_attention_mask, labels=None):
        output = self.model(
            input_ids,
            attention_mask=attention_mask,
            labels=labels,
            decoder_attention_mask=decoder_attention_mask
        )
        return output.loss, output.logits

    def training_step(self, batch, batch_idx):
        loss, _ = self(
            input_ids=batch["text_input_ids"],
            attention_mask=batch["text_attention_mask"],
            decoder_attention_mask=batch["labels_attention_mask"],
            labels=batch["labels"]
        )
        self.log("train_loss", loss, prog_bar=True, logger=True)
        return loss

    def validation_step(self, batch, batch_idx):
        loss, _ = self(
            input_ids=batch["text_input_ids"],
            attention_mask=batch["text_attention_mask"],
            decoder_attention_mask=batch["labels_attention_mask"],
            labels=batch["labels"]
        )
        self.log("val_loss", loss, prog_bar=True, logger=True)
        return loss

    def configure_optimizers(self):
        return AdamW(self.parameters(), lr=0.0001)
```

## Training

```python
from pytorch_lightning.loggers import TensorBoardLogger

# Buat logger untuk pelatihan
logger = TensorBoardLogger("lightning_logs", name="news_summary")
```

```python
data_module = NewsSummaryDataModule(
    train_df,
    test_df,
    tokenizer,
    batch_size=4,
    text_max_token_len=64,  # Sebelumnya 128
    summary_max_token_len=128  # Sebelumnya 512
    )

checkpoint_callback = ModelCheckpoint(
    dirpath="checkpoints", filename="best-checkpoint", save_top_k=1, monitor="val_loss", mode="min"
)

trainer = pl.Trainer(
    logger=logger,
    callbacks=[checkpoint_callback],
    max_epochs=5,
    accelerator="gpu",
    devices=1,
    precision=16,  # Tambahkan ini untuk mixed precision
    enable_progress_bar=True
)

model = NewsSummaryModel()
trainer.fit(model, datamodule=data_module)
```

# Evaluation Metrics

```python
def evaluate_model(test_df, trained_model, tokenizer):
    bleu_scores = []
    rouge_scores = {"rouge1": [], "rouge2": [], "rougeL": []}
    scorer = rouge_scorer.RougeScorer(["rouge1", "rouge2", "rougeL"], use_stemmer=True)

    predictions = []
    for _, row in test_df.iterrows():
        text = row["keywords"]
        reference = row["description"]

        text_encoding = tokenizer(
            text, max_length=128, padding="max_length", truncation=True, return_attention_mask=True, return_tensors="pt"
        )
        device = next(trained_model.model.parameters()).device
        input_ids = text_encoding["input_ids"].to(device)
        attention_mask = text_encoding["attention_mask"].to(device)

        generated_ids = trained_model.model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            max_length=150,
            num_beams=5,
            repetition_penalty=2.5,
            length_penalty=1.0,
            early_stopping=True
        )

        generated_summary = tokenizer.decode(generated_ids[0], skip_special_tokens=True, clean_up_tokenization_spaces=True)
        predictions.append(generated_summary)

        # Calculate BLEU Score
        bleu_scores.append(sentence_bleu([reference.split()], generated_summary.split()))

        # Calculate ROUGE Score
        rouge_result = scorer.score(reference, generated_summary)
        for key in rouge_scores:
            rouge_scores[key].append(rouge_result[key].fmeasure)

    avg_bleu = np.mean(bleu_scores)
    avg_rouge = {key: np.mean(values) for key, values in rouge_scores.items()}

    return predictions, avg_bleu, avg_rouge

predictions, avg_bleu, avg_rouge = evaluate_model(test_df, trained_model, tokenizer)

print("Average BLEU Score:", avg_bleu)
print("Average ROUGE Scores:", avg_rouge)
```

```python
import torch

def summarize_text(text):
    text_encoding = tokenizer(
        text,
        max_length = 512,
        padding = "max_length",
        truncation = True,
        return_attention_mask = True,
        add_special_tokens = True,
        return_tensors = "pt"
    )

    # Move text_encoding tensors to the same device as the model
    device = next(trained_model.model.parameters()).device  # Get model's device
    text_encoding = {k: v.to(device) for k, v in text_encoding.items()}

    generated_ids = trained_model.model.generate(
        input_ids=text_encoding["input_ids"],
        attention_mask=text_encoding["attention_mask"],
        max_length=150,  # Atur panjang maksimal output
        num_beams=5,  # Tingkatkan beam search
        repetition_penalty=3.0,  # Tambahkan penalti untuk pengulangan
        length_penalty=1.5,  # Atur preferensi panjang output
        early_stopping=True
    )


    preds = [tokenizer.decode(gen_id,skip_special_tokens=True,clean_up_tokenization_spaces=True)
             for gen_id in generated_ids]

    return " ".join(preds)
```
