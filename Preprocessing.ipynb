```python
import pandas as pd
import re
import warnings
warnings.filterwarnings('ignore')
```

### Hapus Duplikat

```python
duplicates = df[df.duplicated()]
duplicates
```

```python
df = df.drop_duplicates()
df
```

### Hapus NaN

```python
# Menghitung jumlah NaN di kolom Nama Produk dan Deskripsi
nan_nama_produk = df["Nama Produk"].isna().sum()
nan_deskripsi = df["Deskripsi"].isna().sum()

print("Jumlah NaN di kolom Nama Produk:", nan_nama_produk)
print("Jumlah NaN di kolom Deskripsi:", nan_deskripsi)
```

```python
df = df.dropna(subset=['Deskripsi'])
```

### Cleaning

```python
def clean_description(text):
    patterns = [
        r'(Tinggi|Lingkar|Model Menggunakan|Ukuran Badan Model|Size Chart|Lebar Bahu|model menggunakan).*?(\n|$)',
        r'\b\d+(\.\d+)?\s?cm(m)?\b',
        r'(x\s?)+\d*(\.\d+)?\s?,?',
        r'Toleransi ukuran.*?(\n|$)',
        r'Product Detail.*?:?',
        r'[:=]',
        r'\b(Size|size)\s?[A-Za-z0-9]+\b',
        r'(Kode Artikel|Parent SKU).*?(\n|$)',
        r'\b-?\d+\s?kg\b',
        r'\b-?\d+\b',
        r'\(.*?\)',
        r'\b(S|M|L|XL)\b',
        r'\b(Berat|SNI|No REG|toleransi ukuran)\b',
        r'-',
        r'\n+',
        r'%',
        r'\bM\b',
        r'\bCM\b',
        r'\bXS\b',
        r'\bXXL\b',
        r'\bXL\b',
        r'\bL\b',
        r',',
    ]

    for pattern in patterns:
        text = re.sub(pattern, '', text, flags=re.S)

    text = re.sub(r'\s+', ' ', text).strip()

    return text

df["cleaned_description"] = df["Deskripsi"].apply(clean_description)

df["cleaned_description"]
```

```python
df
```

```python
# Fungsi untuk memisahkan huruf kecil dan kapital yang bertempel
def separate_lower_and_upper_case(text):

    if not isinstance(text, str):
        return text

    text = re.sub(r'([a-z])([A-Z])', r'\1, \2', text)

    return text

df['cleaned_description'] = df['cleaned_description'].apply(separate_lower_and_upper_case)
df['cleaned_description']
```

### Lowercase

```python
df[["Nama Produk", "Kategori", "cleaned_description"]] = df[["Nama Produk", "Kategori", "cleaned_description"]].apply(lambda x: x.str.lower())
```

```python
df
```

```python
nan_deskripsi = df["cleaned_description"].isna().sum()

print("Jumlah NaN di kolom Deskripsi:", nan_deskripsi)
```

```python
df = df.drop(columns=['Deskripsi'])
df
```
